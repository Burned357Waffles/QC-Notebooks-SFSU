```python
# imports, part 3, needed to verify you have all the basic required packages installed
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import plot_histogram, plot_bloch_multivector, plot_distribution, array_to_latex
```

```python
# helper function needed to use plot_bloch_vector() method: converts statevector to 3 values
def statevector_to_bloch_vector(state):
    bloch_x = 2 * np.real(state[0] * np.conj(state[1]))
    bloch_y = 2 * np.imag(state[1] * np.conj(state[0]))
    bloch_z = np.abs(state[0])**2 - np.abs(state[1])**2
    return [bloch_x, bloch_y, bloch_z]

# helper function to plot two bloch spheres side by side
# inputs are two statevectors and two title strings
def plot_2bloch_spheres(state1, title1, state2, title2):
    fig, axs = plt.subplots(1, 2, figsize=(10, 4), subplot_kw={'projection': '3d'})
    
    plot_bloch_vector(statevector_to_bloch_vector(state1), ax=axs[0])
    axs[0].set_title(title1, pad=20)

    plot_bloch_vector(statevector_to_bloch_vector(state2), ax=axs[1])
    axs[1].set_title(title2, pad=20)

    plt.show()
```

```python
def run_job(qc, n_shots = 2**8):
    transpiled_circuit = transpile(qc, sim)
    
    print("Circuit before transpilation")
    display(qc.draw('mpl'))
    print("Circuit after transpilation")
    display(transpiled_circuit.draw('mpl', style="iqp"))
    
    job = sim.run(transpiled_circuit, shots=n_shots)
    result = job.result()
    
    # Get counts and output distribution
    counts = result.get_counts(transpiled_circuit)

    state = result.get_statevector()
    print("Final Statevector:")
    display(state.draw("latex"))

    print("Counts:")
    print(counts)
    
    display(plot_distribution(counts))

    return counts
```

```python
def display_statevector(qc):
    # Get Statevector
    state = Statevector.from_instruction(qc)

    # Display Statevector
    print("\nStatevector:")
    display(state.draw("latex"))

    matrix_form = np.array(state).reshape(-1, 1)
    print("\nStatevector as a matrix:")
    display(array_to_latex(matrix_form))
    
    # Display the Unitary Matrix
    print("Unitary matrix:")
    display(Operator(qc).draw("latex"))

    # Display Bloch Sphere
    display(plot_bloch_multivector(state))
```

### Part 1a: Classical Conditional bit flipping

```python
in_data = [0b000, 0b001, 0b010, 0b011,  0b100, 0b101, 0b110, 0b111]
out_data = []

for vec in in_data:
    print(f"input: {format(vec, '03b')}")

    # first AND vec with 1, 
    # then multiply result by our bit mask to get 0 or 0b110, 
    # then XOR the vec and bit mask to flip 2 MSB
    out_vec = vec ^ ((vec & 1) * 0b110)

    print(f"output: {format(out_vec, '03b')}\n")
    out_data.append(out_vec)
```

### Part 1b: Quantum Conditional bit flipping

```python
def construct_inputs(qc, state, bit_length):
    # find the 1s and apply an X qugate
    for b in range(bit_length):
        if state & (1 << b):
            qc.x(b)
```

```python
NUM_BITS = 3

in_data = [0b000, 0b001, 0b010, 0b011,  0b100, 0b101, 0b110, 0b111]
expected_data = [0b000, 0b111, 0b010, 0b101, 0b100, 0b011, 0b110, 0b001]
in_qc = []

# cx gates with control on LSB, which flip both MSBs
qc_flipper = QuantumCircuit(3, name='qc_flipper')
qc_flipper.cx(0, 1)
qc_flipper.cx(0, 2)

# create quantum circuits for each possible input state
for i in range(len(in_data)):
    in_qc.append(QuantumCircuit(NUM_BITS))

for i in range(len(in_qc)):
    print(f"Input: {format(in_data[i], '03b')}")
    print(f"Expected Output: {format(expected_data[i], '03b')}")

    # create the initial quantum state based on the input
    construct_inputs(in_qc[i], in_data[i], NUM_BITS)
    in_qc[i].barrier(label=f"input: {format(in_data[i], '03b')}")

    # append the flip to the initialized circuit
    in_qc[i] = in_qc[i].compose(qc_flipper, qubits=range(3))

    # display the statevector and circuit
    display_statevector(in_qc[i]) 
    display(in_qc[i].draw('mpl'))

in_qc.clear()
```

#### **What gate do you use in your circuit to test what happens on one qubit but then influences what happens on a different qubit?**

The qugate that I used to check the state of one qubit, then modify another is the CNOT qugate

#### **What type of condition, or relationship, does that gate create between pairs of qubits?**

The CNOT qugate entangles the two qubits. In this circuit, $q_0$ and $q_1$ are entangled and $q_0$ and $q_2$ are entangled.

#### **Describe what happens in your circuit (for one qubit pair) when this condition exists?**

When the CNOT qugate executes, it checks if the control qubit is in the $\ket{1}$ state. If it is, then it will apply an X qugate to the target. If the control qubit is in the $\ket{0}$ state, then it will not apply the X qugate.

### Part 2: Implement “Controlled Swap” using CX gate

```python
# creates a swap between qubits 0 and 1
qc_cswap = QuantumCircuit(3)

qc_cswap.ccx(2, 0, 1)
qc_cswap.ccx(2, 1, 0)
qc_cswap.ccx(2, 0, 1)

in_data = [0b000, 0b001, 0b010, 0b011,  0b100, 0b101, 0b110, 0b111]
expected_data = [0b000, 0b001, 0b010, 0b011, 0b100, 0b110, 0b101, 0b111]
in_qc = []

# create quantum circuits for each possible input state
for i in range(len(in_data)):
    in_qc.append(QuantumCircuit(NUM_BITS))

for i in range(len(in_qc)):
    print(f"Input: {format(in_data[i], '03b')}")
    print(f"Expected Output: {format(expected_data[i], '03b')}")

    # create the initial quantum state based on the input
    construct_inputs(in_qc[i], in_data[i], NUM_BITS)
    in_qc[i].barrier(label=f"input: {format(in_data[i], '03b')}")

    # append the controlled swap to the initialized circuit
    in_qc[i] = in_qc[i].compose(qc_cswap, qubits=range(NUM_BITS))

    # display the statevector and circuit
    display_statevector(in_qc[i]) 
    display(in_qc[i].draw('mpl'))

in_qc.clear()
```

#### **What gate do you use in your circuit to test what happens on one qubit but then influences what happens on a different qubit?**

In this circuit I am using 3 CCNOT (Toffoli) qugates. 

#### **What type of condition, or relationship, does that gate create between pairs of qubits?**

The CCNOT qugate entangles the two qubits. This creates a state of entanglement between all 3 qubits.

#### **Describe what happens in your circuit when this condition exists?**

I first created a swap gate between $q_0$ and $q_1$, then I added an additional control qubit to each of these CX gates so that they only trigger if $q_2$ is 1.  
The swap works as follows (from Wong p.162):  

1. Start with the initial states of qubits 0 and 1:

$$\ket{q_0}\ket{q_1}$$

2. After the first CNOT (0 -> 1)

$$\ket{q_0}\ket{q_0 \oplus q_1}$$

3. After the second CNOT (1 -> 0)

\begin{align*}
\ket{q_0 \oplus q_0 \oplus q_1} \, \ket{q_0 \oplus q_1} 
&= \ket{(q_0 \oplus q_0) \oplus q_1} \, \ket{q_0 \oplus q_1} \\
&= \ket{0 \oplus q_1} \, \ket{q_0 \oplus q_1} \\
&= \ket{q_1} \, \ket{q_0 \oplus q_1}
\end{align*}

4. After the third CNOT (0 -> 1)

\begin{align*}
\ket{q_1} \, \ket{q_0 \oplus q_1 \oplus q_1} 
&= \ket{q_1} \, \ket{q_0 \oplus (q_1 \oplus q_1)} \\
&= \ket{q_1} \, \ket{q_0 \oplus 0} \\
&= \ket{q_1} \, \ket{q_0}
\end{align*}

&emsp; All of this only happens if $q_2$ is 1 because of the extra control placed on it.

### Part 3 Evolving towards a specified solution state

```python
qc3 = QuantumCircuit(2)

# Initial state of superposition
qc3.h(0)
qc3.h(1)
qc3.barrier()

# Solution
qc3.ry(np.pi * 0.5, 0)
qc3.cx(1, 0)

display_statevector(qc3)

qc3.save_statevector()
qc3.measure_all()

run_job(qc3, 1000)
```

#### **In this example we are using the Aer statevector and displaying a distribution of counts. Describe the difference between displaying results as histograms of counts compared to displaying the state vector as in Parts 1 and 2 of this assignment. What information is contained in the counts that is not present in the state vector, and why?**

Displaying just the statevector only gives us the probability amplitude of getting each state after measurement. The histogram of counts gives us a real distribution after running the circuit for $N$ counts, which varys due to the random nature of the quantum circuits. The statevector is unable to show us the counts because it is just a probability.

#### **In this example, how many computational basis states exist in this circuit? In other words, what are the potential outcomes of this circuit?**

There are still 4 computational basis states, but only 2 of them ($\ket{01}$ and $\ket{10}$) have non-zero probabilities.

#### **How would this number of potential outcomes change if, say, we had N=4 qubits, or N=8 qubits?**

If we had 4 or 8 qubits, the number of computational basis states increase to 16 or 256 respectively. The only 2 states that would have the non-zero probabilities would be $\ket{0001}$ and $\ket{0010}$ in the 4 qubit case and $\ket{00000001}$ and $\ket{00000010}$ in the 8 qubit case, assuming that the extra qubits are added on as $q_3$, $q_4$, $q_5$, ..., $q_8$.
